```python
storeData["tarih"] = pd.to_datetime(storeData["tarih"])
day = storeData["tarih"].dt.day
month = storeData["tarih"].dt.month
year = storeData["tarih"].dt.year
```

```python
storeData["day"]=day
storeData["month"]=month
storeData["year"]=year

```

```python
storeData = storeData.drop(columns=["tarih"])
storeData.head()
```

```python
from sklearn import preprocessing
storeData = preprocessing.normalize(storeData)
storeData = pd.DataFrame(storeData, columns=['haftanin_gunu', 'city_code', 'magaza_no', 'magaza_alani',
       'saat', 'nem', 'fis_sayisi','store_avrg_temp', 'outside_temp', 'day', 'month', 'year'])
storeData.head()
```

```python
import seaborn as sns

sns.heatmap(storeData.corr(), annot=True, fmt=".1f")
```
